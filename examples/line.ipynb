```python
import numpy as np
from views_dataviz import Line
```

```python
?Line
```

```python
# Create some mock data.
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)
```

```python
line = (
    Line(
        figsize=(10, 6), 
        labelsize=16,
        bbox_to_anchor=[1.25, 1],
        hide_spines=None
    )
    .add_line(t, data1, twinx=True, linewidth=4, label="Exp", ylabel="exp")
    .add_line(t, data2, label="Sin", ylabel="sin", c="firebrick")
    .set_legend()
)
```

You can access the axes to make further adjustments if required, e.g.:

```python
line.ax2.tick_params(axis="y", which="major", labelsize=18)
line.fig
```

Save with `save`:

```python
line.save("example_line.png")
```
