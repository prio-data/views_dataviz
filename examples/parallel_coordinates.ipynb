```python
import pandas as pd
from views_dataviz import plot_parallel_coordinates
```

```python
mock = pd.DataFrame(
    {
        "model": ["model_a", "model_b", "model_c"],
        "score_a": [0.8, 0.5, 0.2],
        "score_b": [0.3, 0.6, 0.5],
        "score_c": [1.5, 2, 1.4]
    }
).set_index("model")
```

```python
plot_parallel_coordinates(
    mock, 
    ["score_a", "score_b", "score_c"], 
    legend_anchor=(1.2, 1),
    figsize=(10, 5),
    lw=3,
    labelsize=14,
    titlesize=20,
    reverse=["score_a", "score_b", "score_c"],
    linestyles={"model_a": "dotted", "model_c": "dashed"},
    linecolors={"model_b": "grey"},
)
```
